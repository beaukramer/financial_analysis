```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ErgodicProcess(start_wealth, gain, loss, n):
    win = 1 + gain
    loss = 1 - loss
    
    steps = np.random.rand(n)
    
    outcomes = []
    
    end_wealth = start_wealth
    for item in steps:
        if item >= 0.5:
            outcomes.append(end_wealth * win)
            end_wealth = end_wealth * win
        else:
            outcomes.append(end_wealth * loss)
            end_wealth = end_wealth * loss
    return np.array(outcomes)
```

```python
outcomes = np.empty((100,100))
for n in range(100):
    outcomes[n] = ErgodicProcess(start_wealth = 100, gain=0.5, loss=0.4, n=100)
```

```python
plt.plot(np.mean(outcomes,axis=1))
plt.plot(outcomes[0])
```
